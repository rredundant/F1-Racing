```python
import requests
import pandas as pd

drivers = ['alonso','massa','hamilton','badoer']
years = []
years.extend(range(2003, 2019))
results = []

for driver in drivers:
    for year in years:
        api=f'http://ergast.com/api/f1/{year}/drivers/{driver}/status.json'
        resp=requests.get(api).json().get('MRData')
        results.append(resp)
        
```

```python
results
```

```python
status_holder_df = pd.DataFrame.from_records(results)
status_holder_df['driverId'] = status_holder_df.StatusTable.map(lambda x: x.get('driverId'))
status_holder_df['season'] = status_holder_df.StatusTable.map(lambda x: x.get('season'))
status_holder_df.head()
```

```python
def unpack_statuses(status):
    statuses = status.get('Status')
    types = [(x.get('status'), x.get('count')) for x in  statuses]
    return types
    
status_holder_df['statuses'] = status_holder_df.StatusTable.map(unpack_statuses)
```

```python
status_holder_df
```

```python
cleaned_status_df = status_holder_df[['driverId','season','statuses']]
```

```python
from bs4 import BeautifulSoup

obscure = requests.get("http://ergast.com/api/f1/status?limit=140")
soup = BeautifulSoup(obscure.text, "lxml")
```

```python
for row in soup.find('statustable').findAll('status'):
    print(row.attrs.get('statusid')) # properties on the status
    print(row.get_text()) # the name of it
```

```python
for row in soup.find('statustable').findAll('status'):
    status_dict[row.attrs.get('statusid')] = row.get_text()
```
